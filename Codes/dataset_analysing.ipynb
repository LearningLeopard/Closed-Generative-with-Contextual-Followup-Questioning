```python
from datasets import load_from_disk
import numpy as np
import matplotlib.pyplot as plt
```

```python
visa_ds = load_from_disk("../Datasets/Visa_QA_V3/")
```

```python
vote_scores = visa_ds['train']['vote_score']

# Print basic statistics
print(f"Min vote_score: {min(vote_scores)}")
print(f"Max vote_score: {max(vote_scores)}")
print(f"Mean vote_score: {np.mean(vote_scores)}")
print(f"Std vote_score: {np.std(vote_scores)}")

# Plot a histogram to visualize the distribution
plt.figure(figsize=(8, 6))
plt.hist(vote_scores, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title("Distribution of Vote Scores")
plt.xlabel("Vote Score")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```
